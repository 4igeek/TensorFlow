## Linear vs Non-linear activation functions.

In this notebook we're going to look at linear vs non-linear activation functions by re-creating them.

An activation function is a neuronâ€™s gatekeeper: it decides whether or not to activate the neuron, based on its input. It also introduces non-linearity to the network, so the model can learn more complicated patterns and make better predictions.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Let's create some data to play with
A = tf.cast(tf.range(-10,10), tf.float32)
```

# Activation functions

Here is where we declare the various activation functions that we will then plot to further explore.

```python
def sigmoid(x): # Non-linear
  return 1/(1+tf.exp(-x))

def relu(x): # Non-linear
  return tf.maximum(0, x)

def linear(x): # Linear
  return (x)
```

```python
# Create a figure and two subplots side by side
fig, (gr1, gr2, gr3, gr4) = plt.subplots(1, 4, figsize=(12, 4))

# Plot data on the first subplot
gr1.plot(A, 'r-')
gr1.set_title('Original Tensor')

# Plot data on the second subplot
gr2.plot(sigmoid(A), 'r-')
gr2.set_title('"Sigmoid" Tensor')

# Plot data on the second subplot
gr3.plot(relu(A), 'r-')
gr3.set_title('"reLU" Tensor')

# Plot data on the second subplot
gr4.plot(linear(A), 'r-')
gr4.set_title('"Linear" Tensor')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
